```python
import pandas as pd
import textstat
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('../Data/train_essays_v3.csv')

# Applying readability tests to each essay
df['flesch_reading_ease'] = df['text'].apply(textstat.flesch_reading_ease)
df['flesch_kincaid_grade'] = df['text'].apply(textstat.flesch_kincaid_grade)
df['gunning_fog'] = df['text'].apply(textstat.gunning_fog)
df['smog_index'] = df['text'].apply(textstat.smog_index)
df['automated_readability_index'] = df['text'].apply(textstat.automated_readability_index)
df['coleman_liau_index'] = df['text'].apply(textstat.coleman_liau_index)

# Combine scatter plots into one figure
fig, axs = plt.subplots(3, 2)
fig.suptitle('Readability Tests by Type')
axs[0, 0].scatter(df['flesch_reading_ease'], df['generated'])
axs[0, 0].set_title('FRE - General Comprehension')
axs[0, 1].scatter(df['flesch_kincaid_grade'], df['generated'])
axs[0, 1].set_title('FKGL - Educational Grade Level')
axs[1, 0].scatter(df['gunning_fog'], df['generated'])
axs[1, 0].set_title('GFI - Professional Complexity')
axs[1, 1].scatter(df['smog_index'], df['generated'])
axs[1, 1].set_title('SMOG - Health Literacy')
axs[2, 0].scatter(df['automated_readability_index'], df['generated'])
axs[2, 0].set_title('ARI - Character Count Insight')
axs[2, 1].scatter(df['coleman_liau_index'], df['generated'])
axs[2, 1].set_title('CLI - Technical Writing')
fig.savefig('../Charts/Readability_Analysis.png')
fig.tight_layout(h_pad=1)

plt.show()

```
